#####  Q10: Produce a list of facilities with a total revenue less than 1000.

The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */

```python
# Import module
from sqlalchemy import create_engine
import pandas as pd

# Create engine: engine
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')

# Create Query 
sql_query = 'SELECT f.name, SUM(CASE WHEN b.memid =0 THEN b.slots * f.guestcost ELSE b.slots * f.membercost END) AS total_revenue'
sql_query += ' FROM Bookings AS b LEFT JOIN Facilities AS f ON b.facid = f.facid'
sql_query += ' GROUP BY f.name HAVING total_revenue < 1000 ORDER BY total_revenue'

# Execute query and store records in DataFrame: df
df = pd.read_sql_query(sql_query, engine)

# Show results
df
```

#####  Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */

```python
# Create Query 
sql_query =  'SELECT (m2.firstname || " " || m2.surname) as Member, (SELECT m1.firstname || " " || m1.surname '
sql_query += 'FROM Members AS m1 WHERE m1.memid = m2.recommendedby) AS Recommendedby '
sql_query += 'FROM Members as m2 where recommendedby > 0 order by Recommendedby'

# Execute query and store records in DataFrame: df
df = pd.read_sql_query(sql_query, engine)

# Show results
df
```

#####  Q12: Find the facilities with their usage by member, but not guests */

List of Members and Facilities

```python
sql_query = 'SELECT DISTINCT m.firstname || " " || m.surname as mname, f.name '
sql_query += 'FROM Bookings AS b INNER JOIN Facilities AS f ON f.facid = b.facid '
sql_query += 'INNER JOIN Members AS m ON m.memid = b.memid WHERE m.memid != 0 ORDER BY mname, f.name'

# Execute query and store records in DataFrame: df
df = pd.read_sql_query(sql_query, engine)

# Show results
df
```

By Usage

```python
sql_query = 'SELECT f.name as Facility, count(*) as Quantity FROM Bookings AS b  '
sql_query += 'INNER JOIN Facilities AS f ON f.facid = b.facid '
sql_query += 'INNER JOIN Members AS m ON m.memid = b.memid WHERE m.memid != 0 GROUP BY f.name ORDER BY f.name'

# Execute query and store records in DataFrame: df
df = pd.read_sql_query(sql_query, engine)

# Show results
df
```

#####  Q13: Find the facilities usage by month, but not guests */

```python
sql_query = 'SELECT f.name, strftime("%m", b.starttime) as month, count(*) as Quantity FROM Bookings AS b  '
sql_query += 'INNER JOIN Facilities AS f ON f.facid = b.facid '
sql_query += 'INNER JOIN Members AS m ON m.memid = b.memid WHERE m.memid != 0 '
sql_query += 'GROUP BY f.name, month ORDER BY f.name, month'

# Execute query and store records in DataFrame: df
df = pd.read_sql_query(sql_query, engine)

# Show results
df
```
